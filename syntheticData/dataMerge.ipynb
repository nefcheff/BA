### Datenbasis für die Modellvalidierung

```python
import pandas as pd
from IPython.display import display
```

```python
path_Bearbeitungscenter= r"C:\Users\fgottardo\OneDrive - Phoenix Mecano AG (Prod)\Bachelor Arbeit\Bachelor Arbeit\DAG\Validation Daten\Bearbeitungscenter_2.xlsx"   #Datenquelle BDE
path_FA = r"C:\Users\fgottardo\OneDrive - Phoenix Mecano AG (Prod)\Bachelor Arbeit\Bachelor Arbeit\DAG\Validation Daten\BeendeteFAs.xlsx"                           #Datenquelle ERP
```

#### Anzeigen von Header BDE Daten

```python

df_Bearbeitungscenter = pd.read_excel(path_Bearbeitungscenter)
display(df_Bearbeitungscenter.head(0))
```

#### Bereinigen BDE Daten

```python
df_Bearbeitungscenter.drop(columns=['Machine - Workcenter', 'WorkCenter', 'WorkCenter Name', 'Artikel No', 'Artikel Beschr.', 'Finished Quantity', 'Vorgänger Machine - Workcenter', 'Vorgänger WorkCenter', 'Vorgänger WorkCenter Name', 'Vorgänger OperationNo', 'Vorgänger EntryDate' ], inplace=True)
display(df_Bearbeitungscenter.head(0))
```

#### Bereinigte Ansicht der BDE Daten

```python
df_Bearbeitungscenter['Key'] = df_Bearbeitungscenter['Key'].str.split('_').str[0]
display(df_Bearbeitungscenter.head(5))
```

#### Anzeigen von Header ERP Daten

```python
df_FA = pd.read_excel(path_FA, skiprows=2)
display(df_FA.head(0))
```

#### Bereinigen ERP Daten

```python
df_FA.drop(columns=['Beschreibung', 'Herkunftsnr.', 'Arbeitsplannr.', 'Zugewiesene Benutzer-ID', 'Status', 'Suchbegriff', 'Sales Order No.', 'Customer Name'], inplace=True)
display(df_FA.head(0))
```

#### inner join und Ansicht des Datenset

```python
df_merged = pd.merge(df_FA, df_Bearbeitungscenter, left_on='Nr.', right_on='Key', how='inner')
display(df_merged.head(5))
```

```python
df_merged.drop(columns=['Key'], inplace=True)
display(df_merged.head(0))
```

#### Umbenenen der Spaltennamen

```python
df_merged.rename(columns={
    'Nr.' : 'FANummer',
    'Menge' : 'Menge',
    'Startdatum' : 'Startdatum',
    'Enddatum' : 'Enddatum',
    'Fälligkeitsdatum' : 'Fälligkeitsdatum',
    'Fertigstellungsdatum' : 'Fertigstellungsdatum',
    'OperationNo' : 'OperationNo',
    'EntryDate' : 'Logdatum',
    'Durchlauf Zeit (h)' : 'DurchlaufszeitIst(h)',
    'Durchlauf Zeit (t)' : 'DurchlaufszeitIst(t)',
    'Setup And Runtime Zeit (h)' : 'Rüst-Bearbeitungszeit(h)',
    'Setup Zeit(h)' : 'Rüstzeit(h)',
    'Run Zeit(h)' : 'Bearbeitungszeit(h)'
}, inplace=True)

display(df_merged.head(10))
```

#### Dateninfos

```python
df_merged.describe(include='all')
```

#### Kontrolle keine Null values

```python
df_merged.isnull().sum()
```

#### Correlation und Covarianz ausgeben

```python
cols_merged = [
    'Menge',
    'Startdatum',
    'Enddatum',
    'Fälligkeitsdatum',
    'Fertigstellungsdatum',
    'Logdatum',
    'DurchlaufszeitIst(h)',
    'DurchlaufszeitIst(t)',
    'Rüst-Bearbeitungszeit(h)',
    'Rüstzeit(h)',
    'Bearbeitungszeit(h)'
]

corr_matrix = df_merged[cols_merged].corr()
cov_matrix  = df_merged[cols_merged].cov()

print("Korrelationsmatrix:")
display(corr_matrix.round(3))
print("\nKovarianzmatrix:")
display(cov_matrix.round(3))
```
